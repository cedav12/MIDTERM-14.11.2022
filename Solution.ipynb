## 1.1

Is there any company with no difference between close price and adj. close price?

```python
print((STE["Close"] == STE["Adj Close"]).sum() == STE.shape[0])
print((TMUS["Close"] == TMUS["Adj Close"]).sum() == TMUS.shape[0])
print((TTWO["Close"] == TTWO["Adj Close"]).sum() == TTWO.shape[0])
```

Companies TMUS and TTWO have for each observation same close and adjusted close price. This means that the first company STE needed to adjust its closing price from dividents and splits.

## 1.2

Highest and lowest adj close price for each company recorded.

```python
print("STE")
print(STE["Adj Close"].describe())
print("TMUS")
print(TMUS["Adj Close"].describe())
print("TWO")
print(TTWO["Adj Close"].describe())
```

Minimal and maximal closing price could be seen above

## 1.3

Logarhitmic returns.

```python
STE["logreturns"] = np.log(STE['Adj Close']/STE['Adj Close'].shift(1))
print("STE log returns\n",STE["logreturns"])
TMUS["logreturns"] = np.log(TMUS['Adj Close']/TMUS['Adj Close'].shift(1))
print("\nTMUS log returns\n",TMUS["logreturns"])
TTWO["logreturns"] = np.log(TTWO['Adj Close']/TTWO['Adj Close'].shift(1))
print("\nTTWO logreturns\n",TTWO["logreturns"])
```

descriptive statistics of logreturns

```python
print("STE log returns\n")
print(STE["logreturns"].describe())
print("\nTMUS log returns\n")
print(TMUS["logreturns"].describe())
print("\nTTWO log returns\n")
print(TTWO["logreturns"].describe())
```

## 1.4

Which day the company recorded highest gain/loss


Firstly we will change the date from string to datetime and set date as an index.

```python
#Changing the data type
STE["Date"] = pd.to_datetime(STE["Date"])
TMUS["Date"] = pd.to_datetime(TMUS["Date"])
TTWO["Date"] = pd.to_datetime(TTWO["Date"])

#Setting index
STE = STE.set_index("Date")
TMUS =TMUS.set_index("Date")
TTWO = TTWO.set_index("Date")
```

```python
minmax = pd.DataFrame({"maximal logreturn": [STE["logreturns"].idxmax(),TMUS["logreturns"].idxmax(),TTWO["logreturns"].idxmax()],
                      "minimal logreturns" : [STE["logreturns"].idxmin(),TMUS["logreturns"].idxmin(),TTWO["logreturns"].idxmin()]},
                     index = ["STE", "TMUS", "TTWO"])

minmax
```

## 1.5

Average calendar weekly volume

```python
print("STE: ",STE[STE.index.weekday != (5,6)]["Volume"].mean())
print("TMUS: ",TMUS[TMUS.index.weekday != (5,6)]["Volume"].mean())
print("TTWO: ",TTWO[TTWO.index.weekday != (5,6)]["Volume"].mean())
```

## 1.6

The highest total return.

```python
total_returns = {"STE" : STE["logreturns"].sum(),"TMUS" : TMUS["logreturns"].sum(),"TTWO" : TTWO["logreturns"].sum()}
print(total_returns)
```

TMUS made the highest returns
